Risk model answers one yes/no question : is this customer likely to default ? It does by looking at features (income, debt, credit history), assigning weights to them.

We can use logistic regression because its simple, explainable and banks use it

**Logistic Regression takes multiple inputs, combines them and outputs a probability between 0 and 1**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv("/Users/starboy/Documents/Projects/credit_risk_segmentation/data/processed/model_ready_data.csv")
```

```python
features =[
    "debt_to_income",
    "loan_to_income",
    "credit_history_years",
    "annual_income"
]
```

```python
X = df[features]
y = df["target"]
```

**Checking for bad values**

```python
X.isna().sum()
```

```python
np.isinf(X).sum()
```

```python
# Turn "infinite" into "missing"

X = X.replace([np.inf, -np.inf], np.nan)
X = X.fillna(X.median())
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
model = LogisticRegression()
model.fit(X_scaled, y)

coefficients = pd.Series(
    model.coef_[0],
    index=features
).sort_values(ascending=False)

coefficients
```
